```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0,5,10)
x
```

```
y = x**2
y
```

```
plt.plot(x,y)
plt.show() # printing the plot instead of just showing the plot
```

```
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph')
plt.show()
```

```
plt.subplot(1,2,1) # 1st arg - row no.  2ndarg - col no. 3rd arg - plot no.
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
fig = plt.figure()
axes = fig.add_axes([0.5,0.6,0.8,1.0])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Title')
axes.plot(x,y)
plt.show()
```

```
fix,axes = plt.subplots()
```

```
fix,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```
fix,axes = plt.subplots()
axes.plot(x,y)
```

```
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y) 
axes[0].set_title('First')
axes[1].plot(y,x) 
axes[1].set_title('Second')
plt.tight_layout()
plt.show()
```

```
fig = plt.figure(figsize=(3,3))

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y)
```

```
fig, axes = plt.subplots(figsize=(3,3))
axes.plot(x,y)
plt.show()
```

```
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(3,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label='x Square')
axes.plot(x,x**3, label='x Cube')

axes.legend(loc=0)
plt.show()
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Blue',linewidth=2,alpha=0.6) # RGB code can be uses as well, alpha for transparency
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Blue',linewidth=2,linestyle='--',marker='.',markersize=10,markerfacecolor='black')
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Blue',linewidth=2,linestyle='--',marker='.',markersize=10,markerfacecolor='black')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
